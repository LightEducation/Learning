```python
#initialisation
csv = open("employe.csv","w")
csv.write("code,nom,salaire,ville\n")
csv.write("10,Salah Filloul,4000,Casablanca\n")
csv.write("20,Amine Sardi,6000,Oujda\n")
csv.write("30,Soufian Rahal,4000,Taza\n")
csv.write("40,Sara Lafi,3000,Casablanca\n")
csv.close()

#resolution
file = open("employe.csv","r")
lines = file.readlines()
file.close()
print("Les employes:")
for i in range(0,len(lines)):
    print(lines[i].strip())
```

```python
print("Les noms des employes:")
for i in range(1,len(lines)):
    print("\t",i,")",lines[i].split(",")[1])
```

```python
print("Les employes de Casablanca:")
print(lines[0].strip())
for i in range(1,len(lines)):
    if(lines[i].split(",")[3].strip() == "Casablanca"):
        print(lines[i].strip())
```

```python
salair = int(input("Entrer le salair:"))
print("Les employes qui ont le salaire de {} DH".format(salair))
print(lines[0].strip())
for i in range(1,len(lines)):
    if(int(lines[i].split(",")[2]) == salair):
        print(lines[i].strip())
```

# Exercice 7
Utiliser les ensembles pour calculer le score de similarité entre deux phrases avec la méthode du coefficient de Jaccard<br>
Refaire la question précédente pour calculer la similarité entre deux fichiers donnés.

```python
ch1 = "If you want to have big fails achievements in your life, You need to dream bigger."
ch2 = "If you don't know where you're going, any road will get you there."

S1 = set()
S2 = set()
for e in ch1.split(" "):
    S1.add(e)

for e in ch2.split(" "):
    S2.add(e)

coefficient = len(S1 & S2)/len(S1 | S2)
print("le score de similarité entre deux phrases CH1 et CH2 est :",coefficient)
```
